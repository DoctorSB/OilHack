Задание №3. Оценка изменения объема растворенного газа для набора моделей

```python
import os
import pandas as pd
import dateparser
import plotly.graph_objects as go
```

Создание массива значений DISSOLVED для каждой модели

```python
files = os.listdir("input/prts400")
models_dict = {}

for num, file in enumerate(files):
    dissolved_frame = pd.DataFrame(columns=['Date', 'DISSOLVED'])

    with open(f"input/prts400/{file}") as f:
        contents = f.read()  # Reading of text file
        contents = contents.splitlines()  # Split text by rows
        for row in contents:
            if "REPORT" in row:
                date = row.split()[2:5]     # Получение актуальной даты
            if "DISSOLVED" in row:
                name_row = row.split(":")
                name_row = [item.split() for item in name_row]
                value_row = contents[contents.index(row) + 2].split(":")
                value_row = [item.split(".") for item in value_row]

                dissolved_frame.loc[len(dissolved_frame) + 1] = [''.join(date), float(value_row[-2][-3])]
    dissolved_frame['Date'] = dissolved_frame['Date'].apply(lambda x: dateparser.parse(x))
    models_dict[file] = dissolved_frame
```

Визуализация Date VS DISSOLVED для каждой модели

```python
fig = go.Figure()
for model in models_dict:
    fig.add_trace(go.Scatter(name=model,
                             x=models_dict.get(model)['Date'],
                             y=models_dict.get(model)['DISSOLVED'],
                             opacity=0.6,
                             mode="lines",
                             marker_line_width=0.5
                             )
                  )
fig.update_layout(title_text="Оценка изменения объема растворенного газа для набора моделей",
                  title_x=0.15,
                  xaxis_title="Date",
                  yaxis_title="DISSOLVED GAS VOLUME")
fig.show()
```
