Задание №1. Увязка координат скважин с сейсмической сеткой

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

pd.options.mode.chained_assignment = None  # default='warn'
```

Загрузка данных с текстового файла

```python
# Specify path if data file is in different directory from this ipynb file
path_text_file = "input/"  # Example: D:\\bb-python\\output\\location-folder\\
textfile_name = "Extract value^ CUB_FIL4_ANON [Realized] 1; -9"

with open(path_text_file + textfile_name) as f:
    contents = f.read()  # Reading of text file
    contents = contents.splitlines()  # Split text by rows
    contents = [line.split(' ') for line in contents]  # Split every rows by space
    contents = pd.DataFrame(contents, columns=['X', 'Y', 'Z'])  # Wrap the result to dataframe

contents[['X', 'Y', 'Z']] = contents[['X', 'Y', 'Z']].astype(float)
heatmap_frame = pd.pivot_table(contents, index='Y', columns='X', values='Z')  # Preparation data for heatmap
heatmap_frame
```

Загрузка скважин из Excel файла

```python
wells_info_path = "input/"
wells_file_name = 'well_coord.xlsx'

initial_wells_info = pd.read_excel(fr"{wells_info_path + wells_file_name}")
initial_wells_info
```

Поиск ближайшего узла для каждой скважины. (Евклидово расстояние)

```python
wells_info = initial_wells_info.copy()
wells_info[['Y', 'X']] = wells_info[['Y', 'X']].astype(float)
wells_info[['X_nearest', 'Y_nearest']] = None

for num, well in enumerate(wells_info.to_dict('records')):
    contents['X_well'] = well.get('X')
    contents['Y_well'] = well.get('Y')
    contents['Distance'] = np.sqrt(
        (contents['X'] - contents['X_well']) ** 2 + (contents['Y'] - contents['Y_well']) ** 2)
    wells_info['X_nearest'].loc[num] = contents['X'].loc[contents['Distance'].idxmin()]
    wells_info['Y_nearest'].loc[num] = contents['Y'].loc[contents['Distance'].idxmin()]
contents.drop(columns=['X_well', 'Y_well', 'Distance'], inplace=True)
wells_info
```

Построение тепловой карты и обозначение скважин

```python
# Creation of heatmap
fig = go.Figure(data=go.Heatmap(
    x=contents['X'],
    y=contents['Y'],
    z=contents['Z'],
))

# Adding wells position on the heatmap
fig.add_trace(go.Scatter(
    x=wells_info['X'],
    y=wells_info['Y'],
    mode="markers+text",
    marker={'color': 'green',
            'size': 7},
    text=wells_info['well'],
    textposition="top center"
))
fig.update_layout(title_text="Увязка координат скважин с сейсмической сеткой ",
                  title_x=0.5)
fig.show()
```
